```r
# Variables 
n_quads <<- 5
n_times <<- 3 
n_qt <<- n_quads * n_times
n_q <<- 4 
n_samp <<- n_quads * n_q
r_samp <<- 1:n_samp

# Labels 
quad = as.numeric(pData$Quad)
sex = as.numeric(pData$Sex)
lane = as.numeric(pData$Batches)
tlab = c("t1", "t2", "t3")
quads = unique(substr(pData$ID, 1,4))

labels = paste(sapply(1:n_quads, function(i) rep(quads[i],n_times)), tlab  ) 
sexlabels = unlist(lapply(1:n_quads, function(i) rep(unique(cbind(pData$Quad, pData$Sex) )[i,2],n_times)) ) 
timelabels = rep(1:n_times, n_quads)
quadlabels = unlist(lapply(1:n_quads, function(i) rep(unique(cbind(pData$Quad, pData$Sex) )[i,1],n_times)) ) 
get.quad.names <- function(X.cpm.all) {
    quad_names = colnames(X.cpm.all)[seq(0:n_q)*4+1]
    quad_names = sapply(quad_names, function(x) substr(x, 1, 6))
    return(quad_names)
}
```

```r
compute.exp.rank <- function(X.cpm.all) {
    require(listarrays)
    X.rank = lapply(1:n_qt, function(i) t(apply(X.cpm.all[,(1:n_q) + n_q*(i-1) ],  1, rank )))   
    X.rank2 = do.call(cbind, X.rank)
    X.tt1 = X.rank2[,(r_samp)]
    X.tt2 = X.rank2[,(r_samp+n_samp)]
    X.tt3 = X.rank2[,(r_samp+(n_samp*2))]
    
    X.t1 = X.rank2[,-(r_samp)]
    X.t2 = X.rank2[,-(r_samp+n_samp)]
    X.t3 = X.rank2[,-(r_samp+(n_samp*2))]

    X.t1.q = (X.t1[,r_samp] + X.t1[,r_samp+n_samp] )/2
    X.t2.q = (X.t2[,r_samp] + X.t2[,r_samp+n_samp] )/2
    X.t3.q = (X.t3[,r_samp] + X.t3[,r_samp+n_samp] )/2

    X.t1.r = lapply(1:n_quads, function(i) t(apply(X.t1.q[,(1:n_q) + n_q*(i-1) ],  1, rank) ))
    X.t2.r = lapply(1:n_quads, function(i) t(apply(X.t2.q[,(1:n_q) + n_q*(i-1) ],  1, rank) ))
    X.t3.r = lapply(1:n_quads, function(i) t(apply(X.t3.q[,(1:n_q) + n_q*(i-1) ],  1, rank) ))

    X.tr1 = do.call(cbind,X.t1.r)
    X.tr2 = do.call(cbind,X.t2.r)
    X.tr3 = do.call(cbind,X.t3.r)

    colnames(X.tr1) = pData$altID[r_samp]
    colnames(X.tr2) = pData$altID[r_samp]
    colnames(X.tr3) = pData$altID[r_samp]

    rownames(X.tr1) = rownames(X.cpm.all)
    rownames(X.tr2) = rownames(X.cpm.all)
    rownames(X.tr3) = rownames(X.cpm.all)

    rownames(X.tt1) = rownames(X.cpm.all)
    rownames(X.tt2) = rownames(X.cpm.all)
    rownames(X.tt3) = rownames(X.cpm.all)

    test = list(X.tt1=X.tt1, X.tt2=X.tt2, X.tt3=X.tt3)
    training = list(X.tr1=X.tr1, X.tr2=X.tr2, X.tr3=X.tr3)
    return(list(training=training, test=test))
}
```

```r
require(coin)
infected = c()
infected_quad = c()
for (quad in c("12-10", "15-50")) {
    for (ident in 1:4) {
        for (time in 1:3) {
            label = paste0(quad, ident, '_', time)
            if (time > 1)
                infected <- c(infected, label)
            infected_quad <- c(infected_quad, label)
        }
    }
}
infected_index = unlist(sapply(colnames(X.cpm.all), function(x) any(x == infected)))
infected_quad_index = unlist(sapply(colnames(X.cpm.all), function(x) any(x == infected_quad)))
```

```r
# vs all
infected_index = unlist(sapply(colnames(X.cpm.all), function(x) any(x == infected)))
p.values = unlist(apply(X.cpm.all, c(1), function(x) {
    x[is.na(x)] = 0
    df = data.frame(value=as.numeric(c(x[infected_index], x[!infected_index])), 
               state=factor(c(rep('infected', length(which(infected_index))), rep('not', length(which(!infected_index))))))
    if (mean(df$value) == 0)
        return(NA)
    else if (min(df$value) == max(df$value))
        return(1)
    else {
        result = wilcox_test(value ~ state, df, distribution="exact")
        return(pvalue(result))
    }
}))
q.values <- p.adjust(p.values, method = "BH")
write.table(cbind(p.values, q.values), '../data/Hess_2022/deg_infected_vs_all.csv', sep=",")
```

```r
# within quads
infected_index = unlist(sapply(colnames(X.cpm.all), function(x) any(x == infected)))
infected_index = infected_index[infected_quad_index]
                               
print(dim(X.cpm.all[,infected_quad_index]))
                               
p.values = unlist(apply(X.cpm.all[,infected_quad_index], c(1), function(x) {
    x[is.na(x)] = 0
    df = data.frame(value=as.numeric(c(x[infected_index], x[!infected_index])), 
               state=factor(c(rep('infected', length(which(infected_index))), rep('not', length(which(!infected_index))))))
    if (mean(df$value) == 0)
        return(NA)
    else if (min(df$value) == max(df$value)) {
        return(1)
    } else {
        result = wilcox_test(value ~ state, df, distribution="exact")
        return(pvalue(result))
    }
}))
q.values <- p.adjust(p.values, method = "BH")
write.table(cbind(p.values, q.values), '../data/Hess_2022/deg_infected_within_quad.csv', sep=",")
```

```r
require(dplyr)
threshold = 0.05
readDegForInfection <- function(file) {
    df = read.table(file, sep=",")
    df = cbind(df, read.table("../data/Hess_2022/log2FC_deg.csv", sep=","))
    df[,'Name'] = sapply(df[,'Name'], toupper)    
    return(df)
}
deg_mat <- (deg_mat %>% group_by(Gene, Log2FC.infected.vs.uninfected) %>% summarize(Liver.Cell.Group=toString(paste(Liver.Cell.Group, sep=":"))) %>% ungroup())
```

```r
# within_quad (no direction)
file = "../data/Hess_2022/deg_infected_within_quad.csv"
print(length(unique(deg_mat$Gene)))
df = readDegForInfection(file)
df = cbind(df, Row.names=rownames(df))
m = merge(df, deg_mat, by.x='Name', by.y='Gene', all.x=TRUE)
print(dim(m))
# Detected
print(length(which(!is.na(m[,'p.values']) & m[,'Name'] != '')))
m = m[!is.na(m[,'p.values'] & m[,'Name'] != ''),]
print(length(which(!is.na(m[,'Liver.Cell.Group']))))
# Overlapped with DEG
print(length(which(m[,'q.values'] < threshold)))
m = m[m[,'q.values'] < threshold & !is.na(m[,'Liver.Cell.Group']),]
print(dim(m))
# Final summary
print(unique(m[,'Liver.Cell.Group']))
group <- m %>% group_by(Liver.Cell.Group) %>% summarize(count=n()) %>% ungroup()
print(group)
write.table(m, "../data/Hess_2022/overlapped_deg_within_quad.csv", sep=",")
```
